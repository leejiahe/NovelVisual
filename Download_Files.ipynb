## Downloading Github Repo

__It takes around <span style = 'color:red'>5</span> min__

```python
from git import Repo, remote

url = 'https://github.com/MKLab-ITI/image-verification-corpus.git'
path = os.path.join(os.getcwd(), 'MediaEval')

class Progress(remote.RemoteProgress):
    def update(self, op_code, cur_count, max_count=None, message=''):
        print ('update({}, {}, {}, {})'.format(op_code, cur_count, max_count, message))

Repo.clone_from(url, path, progress = Progress())
```
